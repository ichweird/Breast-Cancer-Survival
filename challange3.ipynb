## import libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import metrics
from sklearn.impute import SimpleImputer

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

### import data

```python
database = pd.read_csv('C:/Users/pc/Downloads/1729183122211/challenge 3/data/train.csv')
database.head()
```

### we turn chars into booleen

```python
data = database.copy()
data['Marital_Status'] = data['Marital_Status'].map({'Married':1 , 'single':0})
data['Radiation_Therapy'] = data['Radiation_Therapy'].map({'Yes':1 , 'No':0})
data['Chemotherapy'] = data['Chemotherapy'].map({'Yes':1 , 'No':0})
data['Hormone_Therapy'] = data['Hormone_Therapy'].map({'Yes':1 , 'No':0})

data.head()
```

### we separe the target from features

```python
target = data['Survival_Status']

data.drop(['Survival_Status'],axis=1)

data
```

### scale the data

```python
scaler = StandardScaler()

data_scaled = scaler.fit_transform(data)
data_scaled=pd.DataFrame(data_scaled)
data_scaled
```

### Split Data into Training and Testing Sets

```python
data_train,data_test,target_train,target_test = train_test_split(data_scaled,target,test_size=0.8,random_state=23)
```

lentgh of each part

```python
len(data_test) ,len(data_train)
```

### removing the NaN values 

```python
data_train_clean = data_train.dropna()
target_train_clean = target_train[data_train_clean.index]

data_test_clean = data_test.dropna()
target_test_clean = target_test[data_test_clean.index]

```

number of values without NaN values


```python
len(data_test_clean) ,len(data_train_clean)
```

#### using knn model ^O^

```python
knn = KNeighborsClassifier(n_neighbors=5, p=2)
knn.fit(data_train_clean , target_train_clean)
```

#### Acurracy and F1 score

```python
yknn = knn.predict(data_test_clean)
print('acurracy:', metrics.accuracy_score(target_test_clean , yknn))
print('F1 score:', metrics.f1_score(target_test_clean , yknn ))
```

### creating csv file

```python
submission = pd.DataFrame({
    'Patient_ID': range(len(yknn)),  
    'Predictions': yknn
})
```

```python
submission.to_csv('submission.csv', index=False)
```
